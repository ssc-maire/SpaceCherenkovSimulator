```python
def single_module_analysis(id:str,output_tuple,thresh=20):
  output_tuple.set_default_DF(output_tuple.hits[f'FL{id}_tuple_fluence;1_CK'], thresh=thresh)
  output_tuple.plot_pulse_height_distribution()
  output_tuple.primary_particle_Cherenkov_tuple.plot_pulse_height_distribution()
  phs1 = output_tuple.pulse_height_distribution_monte_carlo.pulse_height_values.reset_index()
  pri1 = output_tuple.hits[f'FL{id}_tuple_fluence;1_PR']
  pri1.hist(column=["theta"], bins=18)
  pri1.hist(column=["phi"], bins=36)
  des = len(output_tuple.pulse_height_distribution_monte_carlo.pulse_height_values)
  print(f'Number of events in FL{id}_CR = {des}')
  print(f'Number of events in FL{id}_PR = {len(pri1)}')
  print(f'The effective area is {des/events*area} cm2 ')
```
